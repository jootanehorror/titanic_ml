```python
def ticketbin(x):
    if x==1:
        return 0
    else:
        return 1
```

```python
def AgeGroup(x):
    if x <=18:
        return 0
    elif  x<=50:
        return 1
    else:
        return 2
```

```python
def name_remake(x):
    return x.split(",")[1].split(" ")[1]
```

```python
def cabinn(x):
    if type(x) is list:
        return x[0][0]
    else:
        return x[0]
```

```python
def pclass1cabin(x):
    a,b,c,d,e=40,114,100,63,56
    if x>a and x<e:
        if x<((a+e)/2):
            return "A"
        else:
            return "E"
    elif x>e and x<d:
        if x<((e+d)/2):
            return "E"
        else:
            return "D"
    elif x>d and x<c:
        if x<((d+c)/2):
            return "D"
        else:
            return "C"
    elif x>c and x<b:
        if x<((c+b)/2):
            return "C"
        else:
            return "B"
    elif x<=a:
        return "A"
    else:
        return "B"
```

```python
def pclass2cabin(x):
    d,e,f=13,12,24
    if x>e and x<d:
        if x<((d+e)/2):
            return "E"
        else:
            return "D"
    elif x>d and x<f:
        if x<((d+f)/2):
            return "D"
        else:
            return "F"
    elif x<=e:
        return "E"
    else:
        return "F"
```

```python
def pclass3cabin(x):
    e,g,f=11,12,14
    if x>e and x<g:
        if x<((g+e)/2):
            return "E"
        else:
            return "G"
    elif x>g and x<f:
        if x<((g+f)/2):
            return "G"
        else:
            return "F"
    elif x<=e:
        return "E"
    else:
        return "F"
```

```python
def cabin_remake(x):
    if x=="A":
        return 0
    elif x=="B":
        return 1
    elif x=="C":
        return 2
    elif x=="D":
        return 3
    elif x=="E":
        return 4
    elif x=="G":
        return 5
    elif x=="F":
        return 6
    elif x=="T":
        return 7
```
